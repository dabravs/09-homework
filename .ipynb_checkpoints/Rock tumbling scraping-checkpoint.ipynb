```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://rocktumbler.com/blog/rock-and-mineral-clubs/")
doc = BeautifulSoup(response.text)
print(doc)
```

```python
#Here we are practicing on one row before we get to our for-loop 
doc.find_all('tr')[2].find_all('td')[0].text #this is how we get to the name of a AL rock and mineral club 
doc.find_all('tr')[2].find_all('td')[1].text #this is the location for an AL rock and mineral club 
```

```python
societies=doc.find_all('tr')
#print(societies)
len(societies)
```

```python
#here is what we were doing that did NOT work
for society in societies[:10]:
    print(society.find_all('td')[0].text)
    print(society.find_all('td')[1].text)
   
```

```python
#this DOES work! 
for society in societies[:10]:
    if society.h3 is not None:
        pass
    else:
        print(society.find_all('td')[0].text)
        print(society.find_all('td')[1].text)
        print('-----')
```
